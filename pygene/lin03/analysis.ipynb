```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
from pathlib import Path
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100

from setup import EQUILIBRIA, PSINORM_VALUES, TIGRESS_DIR, DEFAULTS_DIR, VERSION
%aimport analyze

def get_scans(
    scan_name: str, 
    plot_index: int = None,
) -> dict:
    scans = {}
    for eq in EQUILIBRIA:
        scans[eq] = {}
        for pn in PSINORM_VALUES[1:4]:
            scan_dir = TIGRESS_DIR / eq / pn / scan_name
            assert scan_dir.exists()
            scans[eq][pn] = analyze.LinearScan(scan_dir)
            scans[eq][pn].plot_omega(logx=True, verbose=False)
            if plot_index is not None:
                scans[eq][pn].plot_mode(index=plot_index, fields='apar')
    return scans
```

# nx, nz, hypz scan

```python
nxnzhypz_scans = get_scans('nxnzhypz2')
```
